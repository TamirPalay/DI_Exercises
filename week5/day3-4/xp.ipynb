Data visualisation is important in Data analysis because we can quickly see relationships between variables and come to quick conclusions without having to do many calculations. Line plots usually show a relationship of an independent variable over time


```python
import matplotlib.pyplot as plt

temps=[72, 74, 76, 80, 82, 78, 75]
days=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.plot(days, temps)
plt.ylabel('Temperature F')
plt.xlabel('Day')
plt.title('Daily Temperatures')
plt.show()
```

```python
sales=[5000, 5500, 6200, 7000, 7500]
months=['January', 'February', 'March', 'April', 'May']

plt.bar(months, sales)
plt.ylabel('Sales Amount ($)')
plt.xlabel('Month')
plt.title('Monthly Sales')
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('Student Mental health.csv')
data.head()
```

```python
import seaborn as sns
CGPA=data['What is your CGPA?']
#sort the range ascending
CGPA=CGPA.sort_values()
sns.histplot(CGPA, bins=4, color='green')
plt.xlabel('CGPA')
plt.title('Distribution of CGPA')
plt.ylabel('Frequency')
```

```python

anxiety_by_gender = data.groupby('Choose your gender')['Do you have Anxiety?'].apply(lambda x: (x == 'Yes').mean()).reset_index()
anxiety_by_gender.columns = ['Gender', 'Proportion with Anxiety']

# Create bar plot with Seaborn
sns.barplot(x='Gender', y='Proportion with Anxiety', data=anxiety_by_gender, palette='viridis')
plt.title('Proportion of Students Experiencing Anxiety by Gender')
plt.ylabel('Proportion')
plt.show()
```

```python
# Compute total panic attacks per age
panic_by_age = data.groupby('Age')['Panic_Numeric'].sum().reset_index()
panic_by_age.columns = ['Age', 'Total_Panic_Attacks']

# Create scatter plot
sns.scatterplot(x='Age', y='Total_Panic_Attacks', data=panic_by_age, alpha=0.6, s=50, color='red')
plt.xlabel('Age')
plt.ylabel('Total Panic Attacks')
plt.title('Total Panic Attacks per Age')
plt.show()
```
