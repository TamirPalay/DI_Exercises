# Exercise 3: Identify Data Types

```python
# Exercise 3: Identify Data Types
# Categorise each column as Quantitative or Qualitative

import pandas as pd

sleep_data = pd.read_csv('Time Americans Spend Sleeping.csv')

# Define column classifications
quantitative_cols = []
qualitative_cols = []

for col in sleep_data.columns:
    if sleep_data[col].dtype in ['int64', 'float64']:
        quantitative_cols.append(col)
    else:
        qualitative_cols.append(col)

print("=" * 50)
print("QUANTITATIVE COLUMNS (Numeric / Measurable):")
print("=" * 50)
for col in quantitative_cols:
    print(f"  - {col} (dtype: {sleep_data[col].dtype})")

print()
print("=" * 50)
print("QUALITATIVE COLUMNS (Categorical / Descriptive):")
print("=" * 50)
for col in qualitative_cols:
    print(f"  - {col} (dtype: {sleep_data[col].dtype})")

print()
print(f"Total columns: {len(sleep_data.columns)}")
print(f"  Quantitative: {len(quantitative_cols)}")
print(f"  Qualitative:  {len(qualitative_cols)}")
```

The "How Much Sleep Do Americans Really Get?" dataset contains 9 columns in total, which split evenly into 4 quantitative and 5 qualitative types. The quantitative columns — index, Year, Avg hrs per day sleeping, and Standard Error — are all numeric and measurable: Year and index are integers that simply identify the record chronologically, while Avg hrs per day sleeping is a continuous float representing the core measurement of the dataset, and Standard Error (also a float) quantifies the statistical uncertainty around that estimate. The qualitative columns — Period, Type of Days, Age Group, Activity, and Sex — are all categorical strings (dtype: object) that describe the conditions under which the sleep data was collected. For example, Age Group segments respondents by demographic bracket, Sex distinguishes gender, and Activity and Type of Days narrow down the context of the measurement. Understanding this split matters because quantitative columns are candidates for statistical analysis (means, trends, correlations), whereas qualitative columns are used for grouping, filtering, and segmentation in any deeper analysis.

The Iris dataset contains six columns, five quantitative and one qualitative. The quantitative columns — SepalLengthCm, SepalWidthCm, PetalLengthCm, and PetalWidthCm — are all continuous numerical measurements (stored as float64), meaning they represent real-world physical dimensions that can be measured on a scale and compared mathematically. Id is also flagged as quantitative by dtype (int64), though in practice it's just a row identifier with no analytical meaning — a good reminder that dtype alone doesn't always tell the full story, and context matters. The single qualitative column, Species, is categorical (object dtype): it holds discrete labels — Iris-setosa, Iris-versicolor, Iris-virginica — that group each flower by type rather than measuring anything. You can't average a species name, but you can count it, group by it, and use it as a target label in a classification model, which is exactly what the Iris dataset is famous for.
