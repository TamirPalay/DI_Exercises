```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load an image using PIL
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import rotate

# --- Rotate 30 degrees ---
def rotate_image_30_degrees(image):
    return rotate(np.array(image), 30, reshape=True)

rotated = rotate_image_30_degrees(original_image)  # call the function with ()

# --- Flip horizontally (mirror) ---
flipped_horizontal = ImageOps.mirror(original_image)

# --- Flip vertically ---
flipped_vertical = ImageOps.flip(original_image)

# --- Zoom in 1.2x using resize ---
width, height = original_image.size
zoomed = original_image.resize((int(width * 1.2), int(height * 1.2)))

# --- Display all 4 results ---
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

axes[0].imshow(rotated)
axes[0].set_title("Rotated 30°")

axes[1].imshow(flipped_horizontal)
axes[1].set_title("Flipped Horizontal")

axes[2].imshow(flipped_vertical)
axes[2].set_title("Flipped Vertical")

axes[3].imshow(zoomed)
axes[3].set_title("Zoomed 1.2x")

plt.tight_layout()
plt.show()
```
